```python
#Age Program to convert age to days,mins,secs,hours

age=int(input("Enter your age:"))
con_to_sec=age*365*24*60*60
con_to_min=age*365*24*60
con_to_hr=age*365*24
con_to_days=age*365
print("You have lived {} seconds, {} minutes, {} hours,{} days till date..".format(con_to_sec,con_to_min,con_to_hr,con_to_days))
```

```python
#Convert Age to Days using Fucntion
def getuserinput():
    age = int(input("Enter your age:"))
    return age
    
def ageconversion(age_user):
    con_to_days = age_user * 365
    return con_to_days

def result():
    ageofuser = getuserinput()
    age_to_days = ageconversion(ageofuser)
    return "You have lived {} days till now..".format(age_to_days)

result()
```

```python
# <div class="_1vC4OE _3qQ9m1">₹549</div>   <span class="_35KyD6">
# https://www.flipkart.com/highlander-men-solid-casual-black-shirt/p/itmeh4z5bzuhah6c?pid=SHTEH5HS3YYSKHJJ&marketplace=FLIPKART

import requests
from bs4 import BeautifulSoup

request = requests.get("https://www.flipkart.com/highlander-men-solid-casual-black-shirt/p/itmeh4z5bzuhah6c?pid=SHTEH5HS3YYSKHJJ&marketplace=FLIPKART")
content = request.content

surf = BeautifulSoup(content, "html.parser")

nameOfShirt = surf.find("span",{"class":"_35KyD6"})
print("Name of shirt is:", nameOfShirt.text.strip())

price_finding = surf.find("div", {"class": "_1vC4OE _3qQ9m1"})
print("Cost of shirt is:",price_finding.text.strip())

offer=surf.find("div",{"class":"VGWI6T _1iCvwn _9Z7kX3"})
print("Offer is:",offer.text.strip())

#Converting Price to number
#Removing rupee symbol
price = price_finding.text.strip()
finalPrice = price[1:]
print(float(finalPrice))
print(float(finalPrice) > 500)
```
